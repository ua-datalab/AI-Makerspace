```python
from docling.document_converter import DocumentConverter

source = "bert.pdf"  # document per local path or URL
converter = DocumentConverter()
result = converter.convert(source)
markdown = result.document.export_to_markdown()  # output: "## Docling Technical Report[...]"
```

```python
from IPython.display import display, Markdown
display(Markdown(markdown))
```

```python
from pprint import pprint
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import JsonOutputParser, CommaSeparatedListOutputParser


template = ChatPromptTemplate.from_messages([
	("system", """
  		You are an information extraction AI system. You will look at the following research paper, and answer the questions of the user.
  		""".strip()),
	("human", "Question: {message}\n Paper: {paper}"),
	# ("ai", "...")
]
)

chain = template | llm | CommaSeparatedListOutputParser()

# pprint(chain.invoke({
# 	"paper": markdown,
# 	"message": "Give me a summary of this paper"
# 	}).content)

def extractor(content, command):
	
	response = chain.invoke({
	"paper": content,
	"message": command
	})

	return response

# extractor(markdown, "Tell me the list of authors")
```

```python
markdown.split("\n")[0]
contents = markdown.split("\n", maxsplit=1)[1]
extractor(contents, "Tell me the title of the paper")
```
